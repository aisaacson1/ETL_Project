```python
import pandas as pd
import datetime as dt 
from datetime import datetime, date
import numpy as np
```

```python
border_csv = "Border_Crossing_Entry_Data.csv"

border_df = pd.read_csv(border_csv)
```

```python
border_df.head()
```

```python
border_clean = border_df.rename(columns={"Border":"Country"})

border_clean.head()
```

```python
border_country = border_clean.loc[:,'Country'].str.strip('US-,Border')
```

```python
border_country.head()
```

```python
border_clean.update(border_country)
```

```python
border_clean.head()
```

```python
border_clean.dtypes
```

```python
border_date = border_clean.loc[:,'Date']

border_date = pd.to_datetime(border_date, errors='coerce')

border_date
```

```python
border_date_clean = border_date.dt.date
```

```python
border_clean.update(border_date_clean)
```

```python
border_clean.head()
```

```python
border_cleaner = border_clean.loc[:,'Location'].str.strip('POINT (,)')
```

```python
border_clean.update(border_cleaner)
```

```python
border_location = border_clean["Location"].str.split(" ", n = 1, expand=True)
```

```python
border_clean["Latitude"]= border_location[0]

border_clean["Longitude"] = border_location[1]
```

```python
border_clean.drop(columns =["Location"], inplace=True)
```

```python
border_clean.head()
```

```python
border_clean.to_csv("Clean_Border_Data.csv", index=False, header=True)
```
