```python
# SQL Alchemy
from sqlalchemy import create_engine

# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()

# Set path 
# import sys
# sys.path.append('../../../')

# Config variables
from config import remote_db_endpoint, remote_db_port
from config import remote_gwsis_dbname, remote_gwsis_dbuser, remote_gwsis_dbpwd
from config import local_db_user, local_db_pwd, local_db_endpoint, local_db_port, local_db_name

# Import Pandas
import pandas as pd
```

```python
# Create Engine and Pass in MySQL Connection

# csv files
border_csv = "Border_Crossing_Data/Border_Crossing_Entry_Data.csv"
border_df = pd.read_csv(border_csv)

POE_csv = "Resources/CBP_Encounters_2019_2012.csv"
POE_df = pd.read_csv(POE_csv)

# Cloud MySQL Database Connection on AWS
engine = create_engine(f"mysql://{remote_gwsis_dbuser}:{remote_gwsis_dbpwd}@{remote_db_endpoint}:{remote_db_port}/{remote_gwsis_dbname}")
```

```python
# Create a remote database engine connection
conn = engine.connect()
```

```python
border_df.to_sql(name='border_data', if_exists='append', con=conn, chunksize=500, index=False)
POE_df.to_sql(name='POE_data', if_exists='append', con=conn, chunksize=500, index=False)
```

```python
remote_border_data = pd.read_sql("SELECT * FROM border_data WHERE State='Maine'", conn)
print(len(remote_border_data))
remote_border_data.head(50)
```

```python
remote_POE_data = pd.read_sql("SELECT * FROM POE_data WHERE `POE State`='NY'", conn)
print(len(remote_POE_data))
remote_POE_data.head(50)
```
