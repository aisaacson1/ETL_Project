```python
#Import needed libraries
import pandas as pd
```

```python
#Define locations of datafiles using glob 
POE_data = ('resources/Final COR Port of Entry.xlsx')
```

```python
df_POE = pd.ExcelFile(POE_data)
```

```python
df_POE_2018 = df_POE.parse(2, header=1, skiprows=[2,3,4])
df_POE_2018.drop(df_POE_2018.iloc[:, 14:27], inplace=True, axis=1)
df_POE_2018.drop(df_POE_2018.iloc[:, 0:1], inplace=True, axis=1)
df_POE_2018.dropna(inplace=True)
df_POE_2017 = df_POE.parse(3, header=1, skiprows=[2,3,4])
df_POE_2017.drop(df_POE_2017.iloc[:, 14:27], inplace=True, axis=1)
df_POE_2017.drop(df_POE_2017.iloc[:, 0:1], inplace=True, axis=1)
df_POE_2017.dropna(inplace=True)
df_POE_2016 = df_POE.parse(4, header=1, skiprows=[2,3,4])
df_POE_2016.drop(df_POE_2016.iloc[:, 14:27], inplace=True, axis=1)
df_POE_2016.drop(df_POE_2016.iloc[:, 0:1], inplace=True, axis=1)
df_POE_2016.dropna(inplace=True)
df_POE_2015 = df_POE.parse(5, header=1, skiprows=[2,3,4])
df_POE_2015.drop(df_POE_2015.iloc[:, 14:27], inplace=True, axis=1)
df_POE_2015.drop(df_POE_2015.iloc[:, 0:1], inplace=True, axis=1)
df_POE_2015.dropna(inplace=True)
df_POE_2014 = df_POE.parse(6, header=1, skiprows=[2,3,4])
df_POE_2014.drop(df_POE_2014.iloc[:, 14:27], inplace=True, axis=1)
df_POE_2014.drop(df_POE_2014.iloc[:, 0:1], inplace=True, axis=1)
df_POE_2014.dropna(inplace=True)
df_POE_2013 = df_POE.parse(7, header=1, skiprows=[2,3,4])
df_POE_2013.drop(df_POE_2013.iloc[:, 14:27], inplace=True, axis=1)
df_POE_2013.drop(df_POE_2013.iloc[:, 0:1], inplace=True, axis=1)
df_POE_2013.dropna(inplace=True)
df_POE_2012 = df_POE.parse(8, header=1, skiprows=[2,3,4])
df_POE_2012.drop(df_POE_2012.iloc[:, 14:27], inplace=True, axis=1)
df_POE_2012.drop(df_POE_2012.iloc[:, 0:1], inplace=True, axis=1)
df_POE_2012.dropna(inplace=True)
df_POE_2019 = df_POE.parse(1, header=1, skiprows=[2,3,4])
df_POE_2019.drop(df_POE_2019.iloc[:, 11:26], inplace=True, axis=1)
df_POE_2019.drop(df_POE_2019.iloc[:, 5:6], inplace=True, axis=1)
df_POE_2019.drop(df_POE_2019.iloc[:, 8:9], inplace=True, axis=1)
df_POE_2019.drop(df_POE_2019.iloc[:, 0:1], inplace=True, axis=1)
df_POE_2019.dropna(inplace=True)
df_POE_2017
```

```python
df_2019 = pd.melt(df_POE_2019, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'])
df_2018 = pd.melt(df_POE_2018, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'])
df_2017 = pd.melt(df_POE_2017, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'])
df_2016 = pd.melt(df_POE_2016, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'])
df_2015 = pd.melt(df_POE_2015, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'])
df_2014 = pd.melt(df_POE_2014, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'])
df_2013 = pd.melt(df_POE_2013, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'])
df_2012 = pd.melt(df_POE_2012, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'])
df_2012
```

```python
df_2019 = pd.melt(df_POE_2019, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'], var_name="Date", value_name="Encounters")
df_2019[['POE City','POE State']] = df_2019['Ports of Entry based on I-94 Country of Residence (COR)'].str.split(', ', expand=True)
df_2018 = pd.melt(df_POE_2018, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'], var_name="Date", value_name="Encounters")
df_2018[['POE City','POE State']] = df_2018['Ports of Entry based on I-94 Country of Residence (COR)'].str.split(', ', expand=True)
df_2017 = pd.melt(df_POE_2017, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'], var_name="Date", value_name="Encounters")
df_2017[['POE City','POE State']] = df_2017['Ports of Entry based on I-94 Country of Residence (COR)'].str.split(', ', expand=True)
df_2016 = pd.melt(df_POE_2016, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'], var_name="Date", value_name="Encounters")
df_2016[['POE City','POE State']] = df_2016['Ports of Entry based on I-94 Country of Residence (COR)'].str.split(', ', expand=True)
df_2015 = pd.melt(df_POE_2015, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'], var_name="Date", value_name="Encounters")
df_2015[['POE City','POE State']] = df_2015['Ports of Entry based on I-94 Country of Residence (COR)'].str.split(', ', expand=True)
df_2014 = pd.melt(df_POE_2014, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'], var_name="Date", value_name="Encounters")
df_2014[['POE City','POE State']] = df_2014['Ports of Entry based on I-94 Country of Residence (COR)'].str.split(', ', expand=True)
df_2013 = pd.melt(df_POE_2013, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'], var_name="Date", value_name="Encounters")
df_2013[['POE City','POE State']] = df_2013['Ports of Entry based on I-94 Country of Residence (COR)'].str.split(', ', expand=True)
df_2012 = pd.melt(df_POE_2012, id_vars=['Ports of Entry based on I-94 Country of Residence (COR)'], var_name="Date", value_name="Encounters")
df_2012[['POE City','POE State']] = df_2012['Ports of Entry based on I-94 Country of Residence (COR)'].str.split(', ', expand=True)
df_2013
```

```python
df_all = df_2019.append([df_2018, df_2017, df_2016, df_2015, df_2014, df_2013, df_2012], ignore_index=True)
df_all
```
